```python
import plotly.offline as py
fig = go.Figure(data=go.Heatmap(
                z=z_indices,
                x=c,
                y=t,
                colorscale='YlGnBu',
                hoverongaps = False))

fig.update_layout(title="lexis diagram",
            title_x=0.5,
            xaxis_title = "calendar time",
            yaxis_title = "age")

fig.show()
#py.plot(fig, filename='lexis_diagram_greys.html',auto_open = False)
```
